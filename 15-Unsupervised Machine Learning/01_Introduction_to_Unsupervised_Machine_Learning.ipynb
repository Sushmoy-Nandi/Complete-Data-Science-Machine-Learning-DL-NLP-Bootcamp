# Introduction to Unsupervised Machine Learning


## Summary

* **Supervised Machine Learning** utilizes datasets with both independent (input) and dependent (output) features to solve regression and classification problems.
* **Unsupervised Machine Learning** operates on datasets that have no dependent or output features[cite: 16, 17, 18, 90].
* The primary objective of Unsupervised ML is **clustering**, which groups similar data points together into distinct clusters[cite: 10, 11, 12, 85].
* **Customer segmentation** is a highly effective real-world application of clustering, allowing businesses to group buyers by spending habits[cite: 39, 55, 91].
* Key algorithms include **K-means**, **Hierarchical Clustering**, and **DBScan Clustering**[cite: 61, 62, 63].
* **Silhouette Scoring** is an essential mechanism for validating the accuracy and quality of clustering models[cite: 65, 66, 67, 106].

## Exam Notes

### Silhouette Scoring

**Question**: How do you validate the performance of a clustering model in unsupervised machine learning?

**Answer**: Clustering models are validated using **Silhouette Scoring**[cite: 65, 66, 104]. This is a critical evaluation mechanism used to determine how well the data has been grouped into its respective clusters, making it a "super important" concept for validating model outputs[cite: 66, 67, 106].

## Supervised vs. Unsupervised Machine Learning

### Supervised Machine Learning Recap

* Algorithms such as **Linear Regression**, **Logistic Regression**, **Decision Trees**, **Random Forest**, **XGBoost**, **Gradient Boost**, and **AdaBoost** are prime examples of supervised learning[cite: 2, 3].
* In these models, the dataset contains multiple independent features (inputs such as $f_1, f_2, f_3, f_4$) and precisely one dependent feature (the target output)[cite: 4, 5, 6, 79].
* These datasets are leveraged to solve either **regression** or **classification** problem statements[cite: 7, 8, 73].

### Unsupervised Machine Learning Core Concepts

* Unlike supervised learning, datasets in unsupervised learning completely lack a specific output or target feature[cite: 16, 17, 18, 90].
* The goal shifts to **clustering**, a process where the model naturally groups the data points into similar clusters based on underlying patterns[cite: 10, 11, 12, 23, 85].
* For example, given a dataset containing features like age, years of experience, and salary, the algorithm will detect similarities and group individuals with similar attributes into identical clusters[cite: 14, 15, 26, 27, 28, 88, 89].

## Real-World Use Case: Customer Segmentation

* **Customer segmentation** is a standard real-world application of unsupervised machine learning[cite: 39, 55, 91].
* Businesses can group customer purchasing data using relevant features such as salary and spending score[cite: 42, 48, 96].
* When a new product launches, clustering allows businesses to deploy highly targeted marketing and discount strategies:
  * **Regular buyers** can be identified and offered a **15% discount** to reward brand loyalty[cite: 51, 98].
  * **Occasional buyers** can be segmented and offered a more aggressive **20% discount** to heavily incentivize a new purchase[cite: 53, 54, 107].

## Unsupervised Machine Learning Algorithms Covered

* **K-means Algorithm**: A foundational and highly popular clustering algorithm[cite: 61, 101].
* **Hierarchical Clustering**: Grouping data points to build a hierarchy of clusters[cite: 62, 102].
* **DBScan Clustering**: A density-based clustering algorithm used to find arbitrary-shaped clusters[cite: 63, 103].
* **Silhouette Scoring**: While not an algorithm that clusters data, it is a crucial validation mechanism used to evaluate the accuracy and performance of the preceding clustering models[cite: 65, 66, 67, 104, 106].
