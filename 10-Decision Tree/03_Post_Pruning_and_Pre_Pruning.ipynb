# Decision Trees: Post-Pruning and Pre-Pruning


## Summary

* By default, a decision tree will continuously split until it reaches completely pure **leaf nodes**.


* Allowing a tree to split completely to its maximum depth often leads to **overfitting**, which is characterized by high training accuracy but very low test accuracy.


* **Overfitting** represents a scenario of **low bias and high variance**.


* **Pruning** is the primary technique used to reduce this overfitting, similar to how a gardener trims a plant to maintain its shape and health.


* **Post-pruning** involves building the entire tree first and subsequently cutting back unnecessary branches.


* **Pre-pruning** involves tuning model **hyperparameters** to stop the tree from growing to its full depth during the initial construction phase.



## Post-Pruning

* In **post-pruning**, the algorithm first constructs the complete decision tree to its absolute maximum depth.


* After construction is complete, branches are pruned (cut back) with respect to depth or the majority ratio of output categories.


* For example, if an almost-pure node contains 9 "Yes" instances and only 2 "No" instances, forcing a further split is unnecessary; the branch can simply be pruned at that node to form a final **leaf node** that predicts "Yes".


* Trimming these overly specific branches prevents **overfitting** and helps the model generalize much better to new test data.


* **Important limitation**: Post-pruning is strictly best suited for **smaller datasets**. Building a complete decision tree on millions of records before pruning is highly computationally expensive and time-consuming.



## Pre-Pruning (Hyperparameter Tuning)

* **Pre-pruning** prevents the tree from fully constructing by applying mathematical constraints during the actual building phase.


* This is achieved through **hyperparameter tuning**, which dictates the exact rules for how and when the tree should stop splitting.


* Tools like `GridSearchCV` are often used to find the optimal combination of these parameters.


* Key hyperparameters available in libraries like `scikit-learn` (`sklearn`) to control pre-pruning include:


* **`criterion`**: The function used to measure the quality of a split, such as **Gini**, **entropy**, or **log_loss**.


* **`splitter`**: The strategy used to choose the split at each node, such as "best" or "random".


* **`max_depth`**: The absolute maximum allowed depth of the tree branches.


* **`min_samples_split`**: The minimum number of samples required to split an internal node.


* **`min_samples_leaf`**: The minimum number of samples required to exist at a leaf node.


* **`max_features`**: The maximum number of features to consider when looking for the best split.
