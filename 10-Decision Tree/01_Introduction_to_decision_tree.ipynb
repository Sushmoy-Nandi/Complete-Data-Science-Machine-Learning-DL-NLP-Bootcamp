# Decision Tree Classifier


## Summary

* **Decision Tree Algorithms** are essential machine learning models used for solving both **classification** and **regression** problems.


* They serve as the internal foundational algorithm for **ensemble techniques** like **Random Forest** and **Boosting**.


* The fundamental logic of a decision tree mirrors a nested series of `if-else` conditional programming statements.


* There are two primary variants of decision trees: **CART** (binary splits, used by scikit-learn) and **ID3** (multi-node splits).


* Nodes are categorized as **pure splits** (containing only one class, forming a **leaf node**) or **impure splits** (containing mixed classes, requiring further splitting).


* Mathematical concepts like **Entropy** and **Gini Impurity** are used to mathematically evaluate the purity of a node split.


* **Information Gain** is the metric used to determine the optimal feature to select for the next split.



## Decision Tree Variants

Decision tree creation relies on specific algorithms to determine how branches are formed:

* **CART (Classification and Regression Trees)**: This approach strictly creates **binary splits**. Every node division results in exactly two child branches (e.g., Yes/No or True/False). The popular Python **scikit-learn** library utilizes the **CART** technique.


* **ID3**: Unlike CART, the **ID3** algorithm allows for multiple branches from a single root node. A node can be divided into three or more sub-categories simultaneously.



## Tree Structure and Splitting

The goal of constructing a decision tree is to start at the root and recursively branch down until definitive classifications are reached.

### Pure vs. Impure Splits

* **Pure Split**: A split is considered pure when all the resulting data points belong to a single class (e.g., all instances are "Yes" or all instances are "No"). When a pure split is achieved, it becomes a **leaf node**, representing a final predicted output.


* **Impure Split**: If a resulting node contains a mixture of different classes (e.g., a mix of "Yes" and "No" instances), it is an **impure split**. An impure split indicates that the tree must continue to branch and evaluate other features to reach a pure conclusion.

To construct the tree systematically, mathematical formulas are applied to evaluate the data at each node.
