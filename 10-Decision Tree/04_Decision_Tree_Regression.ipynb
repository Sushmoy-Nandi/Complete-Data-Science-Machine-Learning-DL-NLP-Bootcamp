# Decision Tree Regressor


## Summary

* **Decision Tree Regressors** are utilized to solve regression problem statements where the target output is a **continuous feature**.


* Unlike classification models that rely on **Entropy**, **Gini Impurity**, or **Information Gain**, regressors evaluate splits using **Mean Squared Error (MSE)** and **Variance Reduction**.


* To determine the best split, the algorithm evaluates various continuous thresholds (e.g., $Experience \le 2$ vs. $Experience \le 2.5$) and calculates the **Variance Reduction** for each.


* The threshold that yields the highest **Variance Reduction** is selected as the optimal node split.


* When evaluating new test data, the final predicted output is the **average** of the values contained within the resulting **leaf node**.



## Decision Tree Regressor Concepts

### Regression vs. Classification Splits

In a standard Decision Tree Classifier, algorithms use **Information Gain** alongside purity metrics like **Entropy** and **Gini Impurity** to determine which feature best separates the categorical data.

However, when solving a regression problem, the output feature is a **continuous value** (such as predicting a numerical *Salary* based on *Experience* and *Career Gap*). Because the output is continuous rather than categorical, standard **Information Gain** cannot be used. Instead, the model relies on a mathematical concept called **Variance Reduction** to determine the quality of a split.

### Calculating Variance

To evaluate a split, the algorithm must first calculate the variance of the data points at a given node. This is done using the standard **Mean Squared Error (MSE)** formula:

$$Variance = \frac{1}{n} \sum_{i=1}^{n} (y - \hat{y})^2$$



* $n$ represents the number of elements in the node.


* $y$ represents the actual continuous value of a data point.


* $\hat{y}$ represents the **average** (or mean) of all continuous values within that specific node.



### Calculating Variance Reduction

Once the variance for the **root node** and the resulting **child nodes** is calculated, the algorithm determines the **Variance Reduction** to evaluate if the split is effective. The formula is:

$$Variance Reduction = Variance(Root) - \sum (W_i \times Variance(Child_i))$$



* **Variance(Root)**: The MSE of the parent node before the split.


* $W_i$: The weight or ratio of elements that fall into the child node compared to the parent node (e.g., if a child node receives 4 out of 5 total records, its weight is $4/5$).


* **Variance(Child_i)**: The MSE of the specific child node.



The algorithm compares the **Variance Reduction** across multiple potential splits (e.g., evaluating a left split threshold versus a right split threshold). The split that produces the greatest **Variance Reduction** is chosen as the optimal path for the decision tree.

## Making Predictions on Test Data

Once the **Decision Tree Regressor** is fully constructed, it can be used to predict outcomes for new, unseen test data.

The new data point traverses down the tree by following the established threshold conditions (e.g., moving to a specific branch if its feature value is $\le 2.5$). Upon reaching a terminal **leaf node**, the model calculates the **average** of all the continuous values that reside in that specific leaf. This computed average becomes the final output prediction of the regressor model.
