## Data Visualization With Seaborn

Seaborn is a Python visualization library based on Matplotlib that provides a high-level interface for drawing attractive and informative statistical graphics. Seaborn helps in creating complex visualizations with just a few lines of code. In this lesson, we will cover the basics of Seaborn, including creating various types of plots and customizing them.

### 1.2 Importing Libraries

To use Seaborn, import it alongside Matplotlib (which is required to display plots). The standard alias for Seaborn is `sns`.

````
import seaborn as sns
import matplotlib.pyplot as plt
````

## 2. Working with Data

Seaborn includes various built-in datasets that allow for immediate practice without needing external files.

### 2.1 Loading a Dataset

The `load_dataset` function loads these built-in CSV files directly into a Pandas DataFrame.

**Example: The 'Tips' Dataset**
This dataset records restaurant visits and includes features such as total bill, tip amount, gender, smoker status, day, time, and party size.

````
# Load the built-in 'tips' dataset
tips = sns.load_dataset('tips')

# Display the data
print(tips.head())
````

## 3. Basic Plotting (Relational)

### 3.1 Scatter Plots

A scatter plot visualizes the distribution and relationship between two numerical variables.

**Syntax:**

* `x`: Column name for the x-axis.
* `y`: Column name for the y-axis.
* `data`: The DataFrame containing the data.

**Example:**
Visualizing the relationship between **total bill** and **tip**.

````
# Create a scatter plot
sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.title('Scatter Plot of Total Bill vs Tip')
plt.show()
````

* **Insight**: reveals the correlation between the bill amount and the tip given.

### 3.2 Line Plots

Line plots are useful for visualizing trends in data, often over time or ordered categories.

**Example:**
Visualizing the **total bill** relative to **party size**.

````
# Create a line plot
sns.lineplot(x='size', y='total_bill', data=tips)
plt.title('Line Plot of Total Bill by Size')
plt.show()
````

* **Insight**: Demonstrates the trend of bill amounts as the dining party size increases.

## 4. Categorical Plots

Seaborn provides specialized functions to visualize categorical data, allowing for the comparison of numerical distributions across different categories.

### 4.1 Bar Plots

A bar plot represents an estimate of central tendency (usually the mean) for a numeric variable with the height of each rectangle.

**Example:**
Visualizing the average **total bill** based on **gender**.

````
# Create a bar plot
sns.barplot(x='day', y='total_bill', data=tips)
plt.title('Bar Plot of Total Bill by Day')
plt.show()
````

* **Customization**: The `estimator` parameter can change the measure from mean to median, sum, or standard deviation.

## 5. Distribution and Statistical Plots

These plots are essential for understanding the spread, outliers, and probability distribution of the data.

### 5.1 Box Plots

Box plots display the five-number summary (minimum, Q1, median, Q3, maximum) and are excellent for detecting outliers.

**Example:**
Analyzing the distribution of **total bills** across different **days** of the week.

````
# Create a box plot
sns.boxplot(x='day', y='total_bill', data=tips)
plt.title('Box Plot of Total Bill by Day')
plt.show()
````

* **Interpretation**: The box represents the interquartile range (IQR), the internal line is the median, and points outside the whiskers are outliers.

### 5.2 Violin Plots

Violin plots combine a box plot with a Kernel Density Estimation (KDE) plot, showing the distribution density at different values.

**Example:**
Visualizing the density of **total bills** by **day**.

````
# Create a violin plot
sns.violinplot(x='day', y='total_bill', data=tips)
plt.title('Violin Plot of Total Bill by Day')
plt.show()
````

### 5.3 Histograms

Histograms divide data into bins to visualize the distribution of a single numerical variable.

**Example:**
Checking the distribution of the **total bill**.

````
# Create a histogram with KDE
sns.histplot(tips['total_bill'], kde=True, bins=10)
plt.title('Histogram of Total Bill')
plt.show()
````

* **KDE**: Setting `kde=True` adds a smooth curve representing the probability density function.

````
## KDE Plot
sns.kdeplot(tips['total_bill'],fill=True)
````

## 6. Advanced Multivariate Plotting

Seaborn allows for the exploration of relationships between multiple variables simultaneously.

### 6.1 Pair Plots

A pair plot creates a grid of axes, plotting pairwise relationships for every numerical variable in the dataset.

**Example:**
Visualizing relationships between all numerical features in the 'tips' dataset, colored by gender.

````
# Create a pair plot
sns.pairplot(tips)
````

### 6.2 Heatmaps

Heatmaps visualize data in a matrix format where values are represented by color intensity, commonly used for correlation matrices.

**Example:**
Visualizing the correlation between numerical variables.

````
# Calculate correlation matrix
corr_matrix = tips[['total_bill','tip','size']].corr()

# Create a heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
````

* **Correlation**: Values range from -1 to 1.
* **Annot**: `annot=True` displays the specific correlation values in each cell.

````
import pandas as pd
sales_df=pd.read_csv('sales_data.csv')
sales_df.head()
````

````
## Plot total sales by product
plt.figure(figsize=(10,6))
sns.barplot(x='Product Category',y="Total Revenue",data=sales_df,estimator=sum)
plt.title('Total Sales by Product')
plt.xlabel('Product')
plt.ylabel('Total Sales')
plt.show()
````

````
## Plot total sales by Region
plt.figure(figsize=(10,6))
sns.barplot(x='Region',y="Total Revenue",data=sales_df,estimator=sum)
plt.title('Total Sales by Region')
plt.xlabel('Region')
plt.ylabel('Total Sales')
plt.show()
````

## Summary

* Seaborn is an advanced visualization library that improves upon Matplotlib's defaults.
* It includes built-in datasets (e.g., `tips`, `iris`) useful for practicing data visualization.
* **Relational Plots** (Scatter, Line) visualize relationships between numerical variables.
* **Categorical Plots** (Bar, Count) compare distributions across categories.
* **Distribution Plots** (Box, Violin, Histogram) show how data is spread and identify outliers.
* **Multivariate Plots** (Pair Plot, Heatmap) allow for the analysis of multiple variables and correlations simultaneously.
