# Logistic Regression for Multi-Class Classification: One vs Rest

## Summary

* **Logistic Regression** is naturally designed for **binary classification**.
* When a dataset contains more than two output categories, it becomes a **multi-class classification** problem.
* The **One vs Rest (OvR)** technique allows logistic regression to handle multi-class problems.
* OvR works by creating multiple independent **binary classification models**one for each output category.
* Data labels are transformed using **One-Hot Encoding** to train these models.
* During prediction, the model with the highest predicted **probability** determines the final class.

---

## Transition from Binary to Multi-Class

In binary classification, logistic regression separates data into exactly two classes (e.g., Class A vs Class B).

However, many real-world problems contain more than two categories, such as:

* O1
* O2
* O3

To solve such problems using logistic regression, we use the **One vs Rest** approach.

---

## How One vs Rest Works (Training Phase)

The core idea is to convert a multi-class problem into multiple **binary classification problems**.

If there are three categories (O1, O2, O3), the algorithm creates three models:

* **Model M1**
* **Model M2**
* **Model M3**

### Step 1: One-Hot Encoding

The target variable is converted into one-hot encoded form:

* O1  `[1, 0, 0]`
* O2  `[0, 1, 0]`
* O3  `[0, 0, 1]`

### Step 2: Train Individual Binary Models

Each model focuses on one class versus all others.

* **Model M1 (O1 vs Rest)**  
  Classifies:
  * O1  1  
  * O2 and O3  0  

* **Model M2 (O2 vs Rest)**  
  Classifies:
  * O2  1  
  * O1 and O3  0  

* **Model M3 (O3 vs Rest)**  
  Classifies:
  * O3  1  
  * O1 and O2  0  

Each model independently learns its own decision boundary.

---

## Making Predictions (Testing Phase)

After training, a new data point is passed into **all models simultaneously**.

Each model outputs a probability:

Example:

* M1  0.25  
* M2  0.20  
* M3  0.55  

The algorithm selects the class with the **highest probability**.

In this case:

$$
\max(0.25, 0.20, 0.55) = 0.55
$$

Since M3 produced the highest probability, the final prediction is:

**Category O3**

---

## Why One vs Rest Works

* Each classifier specializes in detecting one class.
* Final classification is determined by probability comparison.
* Works well when classes are clearly separable.

---

## Key Takeaways

* Logistic Regression is inherently binary.
* Multi-class problems require an extension method.
* **One vs Rest** converts one multi-class problem into multiple binary problems.
* Final prediction = class with highest probability.
