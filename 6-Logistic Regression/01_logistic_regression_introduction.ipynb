# Logistic Regression: Binary Classification Intuition

## Summary

* **Logistic Regression** is an algorithm primarily used to solve **binary classification problems**.
* Using **Linear Regression** for classification creates structural issues because it relies on an extending **best fit line**.
* **Outliers** can dramatically distort the best fit line in Linear Regression, causing incorrect predictions.
* Linear Regression can produce unbounded predictions that fall **greater than 1** or **less than 0**.
* Logistic Regression resolves this boundary issue by **squashing** the output to strictly fall between **0 and 1**.

---

## Understanding Binary Classification Problems

A **binary classification problem** involves predicting an output that belongs to exactly two categories.

* Example: Using **study hours** to predict whether a student will **pass or fail** an exam.
* The **Pass** category is labeled as `1`.
* The **Fail** category is labeled as `0`.
* The model takes study hours as input and predicts the probability of passing.

---

## Problems with Linear Regression for Classification

Applying **Linear Regression** to classification introduces two major problems.

### 1. The Negative Impact of Outliers

* Suppose we draw a best fit line for study hours.
* We might set a threshold:

  * Predict `1` if output > 0.5
* If a large **outlier** (e.g., 12 study hours) is introduced, the best fit line shifts.
* This shift can cause previously correct predictions (e.g., a 5-hour pass) to fall below the threshold and become misclassified.

### 2. Unbounded Output Values

* A linear best fit line extends infinitely.
* Predictions can be:
  * Greater than 1
  * Less than 0
* In classification, outputs represent probabilities and must lie strictly between **0 and 1**.
