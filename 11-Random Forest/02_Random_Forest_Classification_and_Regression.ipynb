# Random Forest Classification and Regression

## Summary

* **Random Forest** is an ensemble machine learning algorithm used for both **classification** and **regression** tasks.


* It is a specific application of the **bagging technique**.


* The **base learners** utilized in a random forest are exclusively **decision trees**.


* The algorithm uses **row sampling** and **feature sampling** with **replacement** to generate unique training datasets for each base learner.


* For **classification** problems, the final output is decided by a **majority voting classifier**.


* For **regression** problems, the final output is determined by taking the **average output** across all the decision tree models.


* Using a random forest helps convert a model with **high variance** into a **generalized model** with **low variance**.



## Exam Notes

### Random Forest vs. Decision Tree

**Question**: Why should we use Random Forest instead of a standard Decision Tree? 

**Answer**: By default, a **decision tree** splits its nodes until it reaches its maximum depth, which commonly leads to **overfitting**. Overfitting creates a model with **high training accuracy** (low bias) but **low test accuracy** (high variance). **Random Forest** solves this by training multiple decision trees on different data samples. This ensemble approach maintains the **low bias** but converts the **high variance** into **low variance**, resulting in a highly **generalized model** with better test accuracy. While the **time complexity** required to train multiple trees is higher, the significant improvement in accuracy makes it worthwhile.

## Core Concepts of Random Forest

Random forest takes a dataset with a specific size ($d$) and a certain number of features ($m$) and applies the bagging approach. Unlike generic bagging where you can mix different algorithms, the base learners in this technique are exclusively **decision trees**.

### Sampling Techniques

To ensure each decision tree learns different patterns, the algorithm prepares unique data subsets using two primary sampling methods:

* **Row Sampling**: A subset of rows ($d'$) is extracted from the total dataset, where $d'$ is less than the total number of data points $d$.


* **Feature Sampling**: Instead of feeding all features to a single tree, a subset of features (e.g., F1, F2, F4) is selected.


* Both row and feature sampling are performed **with replacement**. This means that while some rows and features might be repeated, the majority of the data will be uniquely replaced for each model's training set.



### Making Predictions

Once all the decision trees are trained on their respective data samples, the model is ready to evaluate **new test data**.

* **Classification**: Each individual tree predicts a class (e.g., 0 or 1). The algorithm then uses a **majority voting classifier** to count the predictions, and the class with the most votes becomes the final output.


* **Regression**: Because the output is a continuous value, the algorithm gathers the numerical predictions from every tree and calculates the **average output** to serve as the final prediction.
