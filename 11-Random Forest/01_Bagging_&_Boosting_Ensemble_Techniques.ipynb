# Ensemble Techniques: Bagging and Boosting

## Summary

* **Ensemble techniques** involve combining multiple machine learning algorithms to achieve a higher **accuracy** than single models.


* These methods are highly effective and frequently used in competitive platforms like **Kaggle** and various **hackathons**.


* The two primary types of ensemble techniques are **Bagging** and **Boosting**.


* **Bagging** trains multiple **base learners** in **parallel** using different samples of the dataset.


* **Boosting** connects **weak learners sequentially**, where each model attempts to correct the errors made by the previous model.



## Ensemble Techniques Overview

* An **ensemble technique** is a method where multiple machine learning models are combined to make predictions.


* Combining models typically results in a **good amount of accuracy**, making it a preferred approach for complex problem statements.


* These techniques can be applied to solve both **classification** and **regression** problem statements.



## Bagging

* **Bagging** utilizes multiple models known as **base learners**.


* These base learners can be the same machine learning algorithm (like a **Decision Tree**) or different algorithms.


* A critical characteristic of bagging is that all base learners are trained **parallelly**, rather than sequentially.


* Each model is trained on a distinct **sample of the dataset**.


* This sampling process involves **Row Sampling and Feature Sampling with replacement**.


* A prominent example of a bagging algorithm is the **Random Forest**.


* Using **Random Forest** instead of a single Decision Tree helps shift from a model with **high variance** (which is prone to **overfitting**) to a more **generalized model** with **low variance**.



### Aggregation in Bagging

* For **binary or multiclass classification problems**, bagging uses a **majority voting classifier** to determine the final prediction.


* If the maximum number of models vote for '0', the final aggregated output will be '0'.


* For **regression problems**, the final output is determined by calculating the **average** of all the continuous outputs generated by the base learners.



## Boosting

* **Boosting** works by combining multiple **weak learners** together to ultimately form a single **strong learner**.


* Unlike bagging, the weak learners in a boosting setup are connected and trained **sequentially**.



### The Boosting Process

* Initially, the entire dataset is passed to the first model (M1) for training and prediction.


* M1 will inevitably have some **incorrectly predicted records**.


* These incorrect records, along with a subset of other records from the dataset, are then passed to the second model (M2).


* M2 trains on this specific data, and its own incorrect predictions are subsequently passed to the third model (M3).


* This process continues **sequentially** across $n$ number of models.



### Combining Predictions in Boosting

* Because of the sequential training on error subsets, each model acts as an "expert" in a specific domain of the data.


* When evaluating new test data, the predictions from all the models are combined.


* Similar to bagging, the final output is generated using a **majority voting classifier** for classification tasks, or by taking the **average** for regression tasks.



### Boosting Algorithms

* Key algorithms in the boosting family include **AdaBoost**.


* Another major algorithm is **Gradient Boosting**.


* A highly optimized version of this is **XGBoost**, which stands for **Extreme Gradient Boost**.
