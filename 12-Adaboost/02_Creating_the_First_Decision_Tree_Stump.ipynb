# AdaBoost: Creating the First Decision Tree Stump

## Goal
Build the **first weak learner** for AdaBoost by testing a 1-level decision tree (stump) on each feature and choosing the split with the **lowest impurity**.

## What Is a Decision Tree Stump?
A decision tree stump is a tree with:
- only **one split**
- depth = **1**
- two output branches

In AdaBoost, each stump is a weak learner. The model improves by combining many such weak learners.

## Step 1: Create a Stump for Every Feature
For each feature, try a split and measure impurity using:
- **Entropy**
- **Gini Impurity**

Lower impurity means cleaner class separation.

## Example Intuition
Suppose we predict **Credit Card Approval** using two features:
- `Salary`
- `Credit`

### Stump on Salary
Split example: `Salary <= 50k`
- Left branch: mixed labels (high impurity)
- Right branch: still mixed

Result: weak separation.

### Stump on Credit
Split example: `Credit == Good`
- Left branch: mostly or fully one class
- Right branch: much cleaner than Salary split

Result: better separation (lower impurity).

## Step 2: Select the Best First Stump
After evaluating all candidate stumps:
1. Compute impurity score for each stump
2. Compare scores
3. Pick the stump with the **minimum impurity**

In this example, the **Credit** stump is selected as the first weak learner.

## Why This Matters in AdaBoost
The first stump sets the starting point. Next, AdaBoost:
- increases weight on misclassified points
- trains the next stump on reweighted data
- combines stump outputs into a stronger classifier

## Quick Takeaway
AdaBoost starts simple:
- test one split per feature
- choose the purest split
- then iteratively focus on hard-to-classify samples
