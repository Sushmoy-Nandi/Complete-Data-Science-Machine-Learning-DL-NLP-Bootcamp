# AdaBoost Machine Learning Algorithm


## Summary

* **Ensemble techniques** like **bagging** and **boosting** are used to solve both classification and regression problems.


* A fully grown **Decision Tree** often leads to **overfitting**, characterized by high training accuracy and low test accuracy, known as **low bias and high variance**.


* **Random Forest**, a **bagging** technique, utilizes multiple **base learners** (Decision Trees) in parallel to convert **low bias and high variance** into **low bias and low variance**.


* **Boosting** algorithms (like **AdaBoost**, **Gradient Boosting**, and **XGBoost**) connect models sequentially rather than in parallel.


* **AdaBoost** utilizes **weak learners**, specifically **Decision Tree Stumps**, which are trees with a depth of exactly one level.


* Individual **Decision Tree Stumps** typically cause **underfitting** due to limited learning from the training dataset, resulting in **high bias and low variance**.


* By sequentially connecting these **weak learners** and assigning **weights** to them, **AdaBoost** combines them into a **strong learner**.



## Exam Notes

### Decision Trees vs. Random Forests

**Question**: If a Decision Tree model is available, why would you choose to use a Random Forest instead?

**Answer**: A fully deep Decision Tree typically leads to **overfitting**, creating a model with **low bias and high variance**. **Random Forest** solves this by supplying different data samples to multiple **base learners**. By aggregating the outputs, it effectively converts the model's performance to **low bias and low variance**.

## Ensemble Techniques Overview

### Bagging

**Bagging** techniques, such as the **Random Forest Classifier** and **Random Forest Regressor**, use multiple **base learners** trained independently on samples of the dataset. To determine the final prediction, these models aggregate the outputs differently based on the problem type:

* **Classification**: Uses a **majority voting classifier** to determine the final class.


* **Regression**: Calculates the **average** of all numerical outputs from the base learners.



### Boosting

**Boosting** fundamentally differs from bagging by connecting models **sequentially**.

* Instead of independent **base learners**, boosting relies on **weak learners**.


* When a model wrongly predicts certain records, those specific records are passed onward to the next sequential model so it can focus on correcting those errors.



## Deep Dive into AdaBoost

### Decision Tree Stumps

The primary **weak learner** used in **AdaBoost** is the **Decision Tree Stump**.

* A stump is restricted to exactly **one level of depth**.


* Because it is so shallow, a single stump hasn't learned much from the training dataset, inherently leading to **underfitting**.


* This underfitting creates a state of **high bias** (low training accuracy) and **low variance**.



### The AdaBoost Mechanism

Unlike Random Forest, **AdaBoost** does not simply take a majority vote or average. Instead, it assigns specific **weights** to each **weak learner** based on their performance.

The mathematical function defining an **AdaBoost** model aggregates these weighted learners:


$$f = \alpha_1(M_1) + \alpha_2(M_2) + \dots + \alpha_n(M_n)$$

* $M_1, M_2, \dots M_n$ represent the individual **Decision Tree Stumps**.


* $\alpha_1, \alpha_2, \dots \alpha_n$ represent the **weights** assigned to those respective stumps.



Combining these sequentially weighted stumps shifts the collective model from a state of **high bias and low variance** toward becoming a fully capable **strong learner**.

## Mathematics and Weight Updates (Supplemental)

To determine the weight ($\alpha$) for each stump, **AdaBoost** calculates the **Total Error** ($TE$) of the stump's predictions. The performance weight is computed as:


$$\alpha = \frac{1}{2} \ln \left[ \frac{1 - TE}{TE} \right]$$



Once a stump is evaluated, the sample weights ($w$) of the dataset are updated before being passed to the next sequential stump:

* **Correctly Classified Points**: The weight is decreased using the formula $w \cdot e^{-\alpha}$.


* **Incorrectly Classified Points**: The weight is increased using the formula $w \cdot e^{\alpha}$.



After updating, the weights are **normalized**, and **bins** are assigned to randomly select and sample the heavily weighted (previously incorrect) data points to effectively train the next stump in the sequence.
