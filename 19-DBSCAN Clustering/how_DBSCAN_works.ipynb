# DBSCAN Clustering

## Summary

* **DBSCAN clustering** is a density-based algorithm that groups data points based on their concentration in various regions.
* The algorithm relies on two primary hyperparameters: **`minPts`** (minimum points) and **epsilon** (radius).
* Data points are classified into three distinct categories: **Core points**, **Border points**, and **Outliers** (or **Noise**).
* A major advantage of **DBSCAN** is its ability to effectively handle **noise** and identify **non-linearly separable clusters**.

## Introduction to DBSCAN

**DBSCAN** is a clustering technique utilized in machine learning, offering an alternative to traditional methods like **K-means** or hierarchical clustering. Instead of relying on distance to a centroid, it groups data based on the density of points in specific regions.

### Key Hyperparameters

To configure the clustering behavior, two main hyperparameters are utilized:
* **Epsilon**: The radius used to draw a circular boundary around a specific data point.
* **Minimum Points (`minPts`)**: The minimum threshold of data points required within the epsilon radius to define a dense region. For example, `minPts` can be set to 4.

*Note: Techniques such as silhouette scoring can be employed to evaluate and select the optimal values for these hyperparameters.*

## Point Classifications



Based on the density of the surrounding area, **DBSCAN** classifies every data point into one of three categories:

### 1. Core Points
* A point is classified as a **Core point** if the number of points within its epsilon radius is greater than or equal to the **`minPts`** threshold.
* This count includes the data point itself.
* Because these dense points are reachable from one another, they combine to form a single cluster.

### 2. Border Points
* A point is considered a **Border point** if the number of data points within its epsilon radius is less than the **`minPts`** requirement.
* While they do not have enough surrounding points to be core points, they are still reachable from a core point and thus belong to the same cluster.

### 3. Outliers (Noise)
* A point is identified as an **Outlier** or **Noise** if it falls outside the dense regions. 
* These points are neither core points nor directly reachable from any core points.
* When a circle is drawn using the epsilon radius around an outlier, no other points will exist within it.

## Advantages of DBSCAN

* **Robust to Noise**: **DBSCAN** is highly effective at identifying and isolating **outliers**, ensuring that noise is separated and skipped during the clustering process.
* **Handles Non-Linear Clustering**: Unlike traditional clustering methods (like **K-means** or Gaussian Mixture EM) that may not account for multi-dimensionality, **DBSCAN** can contort into different shapes to find complex, **non-linearly separable clusters**.
