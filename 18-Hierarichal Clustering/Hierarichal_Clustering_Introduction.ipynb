# Hierarchical Clustering

## Summary

* **Hierarchical clustering** groups data points together without using cluster **centroids**, which is a key difference from K-means clustering.
* The process is split into two primary methods: **Agglomerative** (a bottom-up approach of combining points) and **Divisive** (a top-down approach of dividing clusters).
* **Agglomerative clustering** begins by treating every single data point as an isolated cluster, then iteratively merges the closest pairs until only one large cluster remains.
* A **Dendrogram** is a diagrammatic tool used to track these merges and determine the optimal **number of clusters** ($k$).
* The y-axis of a dendrogram represents the **Euclidean distance** between grouped points.
* To find the best $k$ value, you must find the **longest vertical line such that no horizontal line passes through it**, and draw a horizontal threshold line across it.

## Exam Notes

### Interview Question: K-Means vs. Hierarchical

**Question**: What is the primary architectural difference between K-means clustering and hierarchical clustering?

**Answer**: In K-means clustering, each cluster is defined by a specific **centroid**. In **hierarchical clustering**, the clusters are formed naturally based on distances, and there are **no centroids** utilized. 

### Interview Question: Clustering Approaches

**Question**: How do Agglomerative and Divisive clustering differ in their approach to grouping data?

**Answer**: **Agglomerative clustering** uses a bottom-up approach, starting with individual data points and continually combining them until one large cluster is formed. **Divisive clustering** is the exact reverse; it uses a top-down approach, starting with all data in a single cluster and dividing it down into individual points.

## Introduction to Hierarchical Clustering

The primary goal of clustering algorithms is to take a dataset and organize it into distinct groups. While familiar algorithms like K-means rely on specific **centroids** for each group, **hierarchical clustering** achieves this separation without using centroids at all.



## Types of Hierarchical Clustering

There are two fundamental approaches to hierarchical clustering:

* **Agglomerative Clustering**: This translates to "combining." It is a bottom-up methodology where you begin with the smallest possible elements and continuously group them together.
* **Divisive Clustering**: This translates to "dividing." It is a top-down methodology where you begin with one massive group and iteratively break it apart.

If you understand the geometric intuition of one, you can easily understand the other, as they are direct reverses of each other.



## Steps for Agglomerative Clustering

The geometric intuition for agglomerative clustering relies heavily on calculating distances (such as **Euclidean distance** or Manhattan distance) to group the closest items. The step-by-step process is as follows:

* **Initialize**: Initially, consider each individual data point as its own separate cluster. For example, if you have six points, you start with six distinct clusters.
* **Combine Nearest Points**: Measure the distances to find the two nearest points or clusters, and merge them to create a new, larger cluster.
* **Repeat**: Keep applying this exact same process—finding the next nearest points or clusters and grouping them—until every single point is contained within one single, massive cluster.

## Dendrograms and Choosing the Number of Clusters

Following the agglomerative steps will leave you with a single cluster containing all data. To decide how many functional clusters ($k$) you should actually extract from your data, you must use a visual tool called a **dendrogram**.



### Constructing a Dendrogram

* The x-axis of the dendrogram lists all the original data points.
* The y-axis represents the **Euclidean distance**.
* As you combine clusters in the agglomerative steps, you draw vertical lines upward on the dendrogram to a height matching the Euclidean distance between those two groups, connecting them with a horizontal line.

### Setting the Threshold for $k$

The optimal number of clusters is determined by establishing an **Euclidean distance threshold**. 

* Establishing a threshold involves drawing a continuous horizontal line across the entire dendrogram at a specific y-axis value.
* Count how many vertical lines this new horizontal line intersects. If it passes through two vertical lines, then your number of clusters is two ($k=2$). If it passes through four lines, then $k=4$.
* As a general rule, decreasing the distance threshold directly increases the number of clusters.

### The Longest Vertical Line Hack

Selecting an arbitrary threshold can be difficult, but there is a reliable geometric hack to find the best separation:

* Scan the dendrogram to find the **longest vertical line such that no horizontal line passes through it**. Ensure no other cluster's horizontal connecting lines bisect this vertical span.
* Once you identify this specific vertical stretch, draw your horizontal threshold line directly through it.
* Count the intersections to find a structurally sound $k$ value that mathematically represents the best grouping for your data.
