# Feature Selection vs. Feature Extraction

## Summary

* **Dimensionality Reduction** is essential for managing datasets with a large number of features.
* The three main reasons to perform dimensionality reduction are to **prevent the curse of dimensionality**, **improve model performance**, and **visualize the data**.
* **Feature Selection** involves choosing the most important existing features and dropping those with little to no relationship to the output variable.
* Statistical methods like **Covariance** and **Pearson Correlation** quantify the linear relationship between independent and dependent variables to aid in feature selection.
* **Feature Extraction** is a transformation process where new features are derived from existing, highly important features, reducing dimensions while retaining the essence of the data.

## Exam Notes

### Reasons for Dimensionality Reduction

**Question**: Why should we perform dimensionality reduction? (Common interview question)

**Answer**: There are three primary reasons:
1. To **prevent the curse of dimensionality**.
2. To **improve the performance of the model** by reducing the computational overhead of training across too many dimensions.
3. To **visualize and understand the data** by reducing high-dimensional data into human-readable **2D** or **3D** formats.

## Dimensionality Reduction Overview

When dealing with a dataset containing hundreds of features (dimensions), training a machine learning model becomes computationally expensive and mathematically complex. Applying **Dimensionality Reduction** helps streamline this process. Because humans can only visualize data up to **3D**, reducing dimensions down to **2D** or **3D** allows data scientists to accurately **visualize the data** and deeply understand underlying patterns.

Two core techniques are used to achieve this: **Feature Selection** and **Feature Extraction**.

## Feature Selection

**Feature Selection** is the process of selecting the most crucial independent features that effectively predict the output (dependent feature), while discarding irrelevant ones.

### Understanding Relationships

To select features, we analyze the relationship between an independent feature ($X$) and a dependent feature ($Y$):
* **Positive Linear Relationship**: As $X$ increases, $Y$ increases.
* **Inverse (Negative) Linear Relationship**: As $X$ increases, $Y$ decreases.
* **No Relationship**: $X$ has no predictable impact on $Y$ (scatter plot shows no clear pattern).

### Covariance and Correlation

We quantify these relationships using statistical measurements:

* **Covariance**: Measures the directional relationship between variables.
  $Cov(x,y) = \frac{\sum_{i=1}^{n} (x_i - \bar{x})(y_i - \bar{y})}{n-1}$
  * A **positive value** indicates a positive linear relationship.
  * A **negative value** indicates an inverse linear relationship.
  * A value **approximately equal to zero** indicates no relationship.

* **Pearson Correlation**: Standardizes covariance to provide a value strictly between **-1 and +1**. 
  * Values closer to **+1** indicate a strong **positive correlation**.
  * Values closer to **-1** indicate a strong **negative correlation**.
  * Values near **0** indicate **no relationship**.

### Practical Example: Housing Data

Consider a housing dataset where we want to predict the **Price** of a house.
* **House Size**: Shows a strong linear relationship (high positive covariance/correlation) with the price. This is an **important feature** and should be selected.
* **Fountain Size**: A scatter plot reveals that as fountain size increases, the price remains relatively stagnant. The correlation is near zero. Since this feature lacks a meaningful relationship with the output, it should be **dropped**.

## Feature Extraction

**Feature Extraction** is used when you have multiple independent features that are *all* highly correlated with the output, meaning none can be easily dropped without losing significant predictive power.

### Deriving New Features

Instead of simply dropping variables, Feature Extraction applies a **transformation** to derive a brand **new feature** that captures the essence of the original variables. 

* **Example**: If a dataset contains **Room Size** and **Number of Rooms** as independent features, both are highly important for predicting the house's price. 
* By applying feature extraction, we can compress these two dimensions into a single derived feature, such as **House Size**. 
* While a tiny amount of granular information might be lost, the new **House Size** feature retains enough variance to accurately predict the price.

This allows the model to reduce dimensions (e.g., from 15 features down to 3) while maintaining its predictive accuracy and making the data much easier to visualize.
