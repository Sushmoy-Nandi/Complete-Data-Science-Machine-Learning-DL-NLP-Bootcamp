# Principal Component Analysis (PCA) and Dimensionality Reduction


## Summary

* Introduction to **Principal Component Analysis (PCA)**, which is also commonly referred to as **Dimensionality Reduction**[cite: 1, 2].
* Explanation of the **Curse of Dimensionality**, a phenomenon where continuously adding features eventually causes a model's performance to degrade[cite: 4, 137].
* Two primary methods exist to resolve the curse of dimensionality: **Feature Selection** and **Feature Extraction**[cite: 143].
* **Feature Extraction** involves deriving entirely new, essential features from the original dataset while maintaining the core information[cite: 126, 127].
* The mathematical foundation of PCA relies on extracting **Eigenvectors** and **Eigenvalues** from a **Covariance Matrix** to find paths of maximum variance[cite: 381, 397].

## The Curse of Dimensionality

### Understanding Dimensions
* In machine learning, when we talk about **dimensions**, we are referring directly to the **features** present in a dataset[cite: 13, 14].
* A model trained with a small number of highly relevant features will generally see its accuracy increase as important data points are added[cite: 40, 41]. 
* However, as the number of **features** increases significantly, the model attempts to learn from all of them, leading to **overfitting**[cite: 64, 71, 72].
* **Mathematical confusion**: Because machine learning models are fundamentally mathematical equations, feeding them irrelevant or minimally important features creates internal confusion and unnecessary computational overhead, causing the **model performance to degrade**[cite: 70, 71, 77].

### The Housing Price Example
* Consider an expert trying to predict a **house's price**. Basic features like **house size** and **number of bedrooms** are highly relevant and make prediction easy[cite: 20, 21].
* If you start adding niche parameters—such as proximity to a beach, distance to a celebrity's house, or the exact number of nearby grocery stores—the prediction becomes overly complicated[cite: 92, 96, 98].
* Just as continuously adding irrelevant variables overwhelms a human domain expert and reduces their predictive accuracy, it similarly degrades a machine learning model's performance[cite: 107, 108, 109].

## Solutions to the Curse of Dimensionality

To prevent the degradation of model performance, we must reduce the dataset's dimensions[cite: 112]. There are two main approaches:

### Feature Selection
* This technique involves evaluating the dataset and retaining only the **most important features** to train the model[cite: 115, 123].
* Features that are irrelevant or highly correlated are simply dropped from the process[cite: 259, 260].

### Feature Extraction
* Instead of merely dropping existing features, this process derives **new features** (e.g., $D_1, D_2$) from the original feature set (e.g., $f_1, f_2, f_3$)[cite: 126, 131].
* These newly derived features capture the **essence and variance** of the original data while existing in a lower-dimensional space[cite: 148].
* **Dimensionality Reduction** algorithms, specifically **Principal Component Analysis (PCA)**, utilize this feature extraction methodology[cite: 118, 120, 145].

## Mathematical Intuition Behind PCA

* **PCA** aims to reduce dimensions (e.g., projecting from 2D to 1D, or 3D to 2D) while minimizing the **loss of information**[cite: 148, 294, 301].
* It transforms the original data structure to find the **best Principal Components (PCs)** that capture the **maximum variance** (or spread) of the data[cite: 309, 315, 323].
* The core mathematical operation involves calculating the **Covariance Matrix** of the features to understand their relationships[cite: 381].
* Through **Eigen decomposition**, the algorithm calculates **Eigenvectors and Eigenvalues** ($A v = \lambda v$) from the covariance matrix[cite: 380, 397].
* The magnitude of the **Eigenvalue** determines the significance of its corresponding **Eigenvector**; the vector with the highest magnitude captures the maximum variance and becomes the first **Principal Component**[cite: 390, 392, 393].
