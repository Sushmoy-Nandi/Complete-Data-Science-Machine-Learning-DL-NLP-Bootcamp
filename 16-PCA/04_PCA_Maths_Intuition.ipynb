# Math Intuition Behind Principal Component Analysis (PCA)

## Summary

* The primary goal of the **Principal Component Analysis (PCA)** algorithm is to find the best **Principal Component** line that captures the **maximum variance** after data transformation.
* PCA determines the best principal component using two main mathematical concepts: **Projections** and a **Cost Function** based on variance.
* **Projections** involve mapping original data points (vectors) onto a **unit vector** using a **dot product**, which results in a **scalar value** representing the distance on the new axis.
* The **Cost Function** calculates the variance of these projected scalar values, aiming to find the unit vector that produces the **maximum variance**.
* Instead of trial and error, PCA uses **Eigen decomposition** on a **covariance matrix** to efficiently calculate the optimal unit vector.
* The **Eigenvector** with the highest **Eigenvalue** (magnitude) is selected as the primary Principal Component because it mathematically guarantees the capture of the maximum variance.

## The Goal of PCA

When reducing dimensions (e.g., converting from 2D to 1D), the final goal of the **PCA algorithm** is to identify the best **Principal Component** line. This line is considered the "best" when, after projecting the original data points onto it, the **maximum variance** (or spread) of the data is captured. 

To systematically find this optimal line without relying on trial and error, PCA utilizes two core mathematical steps: **Projections** and a **Cost Function**.

## Projections in PCA

To evaluate how well a new axis captures variance, the original data points must be projected onto it. 


Let a data point be represented as a vector $P_{1}$ with coordinates $(x_1, y_1)$, and let the new proposed axis be represented by a **unit vector** $u$.
When we project vector $P_{1}$ onto the unit vector $u$, we create a new projected point, $P_{1}'$. 

The mathematical equation for this projection is:
$$P_{1}' = \frac{P_{1} \cdot u}{||u||}$$

Because $u$ is a **unit vector**, its magnitude is exactly 1 ($||u|| = 1$). Therefore, the projection equation simplifies to a simple **dot product**:
$$P_{1}' = P_{1} \cdot u$$

By calculating the **dot product** of the original data point coordinates and the unit vector coordinates, we yield a **scalar value**. This scalar value represents the exact distance from the origin to the new projected point $P_{1}'$ along the new axis.

## The Cost Function (Maximizing Variance)

After projecting all original data points ($P_1, P_2, P_3... P_n$) onto the unit vector, we obtain a sequence of projected **scalar values** ($x_1', x_2', x_3'... x_n'$). 

PCA then applies a **Cost Function** to evaluate the quality of this specific unit vector. The cost function simply calculates the **variance** of these newly projected scalar points. 

The variance equation used is:
$$\frac{\sum_{i=1}^{n} (x_i' - \bar{x'})^2}{n}$$

The algorithm's objective is to optimize this **Cost Function** by finding the specific unit vector that results in the **maximum variance**.

## Eigen Decomposition

Continuously guessing and checking different unit vectors to find the maximum variance is computationally inefficient. Instead, PCA uses a linear algebra technique called **Eigen decomposition** to find the exact optimal vectors directly.

The process follows these steps:
1. **Covariance Matrix**: First, calculate the **covariance matrix** between the original features. This matrix represents the relationships and spread of the existing data.
2. **Linear Transformation**: Apply a linear transformation to the covariance matrix to find its **Eigenvectors** ($v$) and **Eigenvalues** ($\lambda$). The governing equation for this transformation is:
   $$Av = \lambda v$$
3. **Select Maximum Magnitude**: The resulting **Eigenvalue** ($\lambda$) represents the **magnitude** of its corresponding **Eigenvector**. 
4. The **Eigenvector** with the highest **Eigenvalue** is selected as **Principal Component 1 (PC1)**. It is mathematically proven that projecting data onto this specific Eigenvector will perfectly capture the **maximum variance**.
